# Heart Disease Prediction

Import libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
```

Load the dataset

Adjust the path to your CSV file

```python
path = r"C:\Users\QFAROOQU\Downloads\Heart_Disease_Prediction.csv"
df = pd.read_csv(path)
```

Define features and target

```python
X = df.drop(columns=['Heart Disease'])
y = df['Heart Disease']
```

Split the data

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2020)
```

Feature Scaling

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```

Initialize and train the SVM model

```python
classifier = SVC(kernel='rbf')  # Using RBF kernel as an example
classifier.fit(X_train_sc, y_train)
```

Evaluate the model

```python
print("Model accuracy:", classifier.score(X_test_sc, y_test))
```

Predicting for a new patient

Create a DataFrame for the new patient with the same feature names

```python
patient_data = {
    'Age': [55], 'Sex': [1], 'Chest pain type': [3], 'BP': [140], 'Cholesterol': [240],
    'FBS over 120': [0], 'EKG results': [1], 'Max HR': [150], 'Exercise angina': [0],
    'ST depression': [1.4], 'Slope of ST': [2], 'Number of vessels fluro': [0], 'Thallium': [2]
}
patient1 = pd.DataFrame(patient_data, columns=X.columns)
```

Scale the patient data

```python
patient1_sc = sc.transform(patient1)
```

Predict

```python
pred = classifier.predict(patient1_sc)
```

Decode the prediction

```python
pred_label = 'Presence' if pred[0] == 'Presence' else 'Absence'
print(f'Prediction for patient: {pred_label}')
```
